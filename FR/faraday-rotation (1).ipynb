This python notebook shows how a linearly polarized light gets rotated when travelling through Interstellar medium

Importing the required libraries

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Declaring the variables : theta, charge of an electron (e), magnetic field in the ISM (B), the electron density (n), mass of an electron (m), speed of light (c), pi, frquency of the emitted light (f) 

```python
theta_max = 8 * np.pi
theta = np.linspace(0, theta_max, 1000)

x = theta
z =  np.sin(theta)
y =  np.cos(theta)
e=4.8*(10**-10) #Charge of an electron in e.s.u
B=10**-6 #Magnetic field in gauss
n1=12.9 #in cm^-3
m=9.1*(10**-28) #mass in grams
c=3*(10**10)
pi=3.14
f=400*(10**6)
```

The distance d in parsec. This variable can be edited to see different roations in the linearly polarized light

```python
# begin edit 

d=3

#end edit
```

```python
n = 1000
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
#Calculating

num=(e**3)*(B)*n1*(10**18)*d
deno=(m**2)*(c**2)*2*pi*(f**2)
phirad=num/deno
z1 =  np.sin(theta+(phirad))
y1 =  np.cos(theta+(phirad))
y11=2*y
y2=y+y1
z2=z-z1
lp1=np.sqrt(np.square(y2)+np.square(z2))

plt.xlabel(r"$\dot{\Theta}$")


ax.plot(x, y, z, 'b', lw=2, label="RCP")
ax.plot(x, y1, -z1, 'r', lw=2,label="LCP")
ax.plot(x, y2, z2, 'k', lw=2,label="LP")
plt.legend(loc='upper left', shadow=True)

plt.show()
```
